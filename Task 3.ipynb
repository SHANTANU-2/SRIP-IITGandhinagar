# MODEL HYPERTUNNIG
implementing hypertunning for Random forest, Linear regresion

```python
model_prams = {
    'RFR' :{
        'model' : RandomForestRegressor(),
        'pram' : {'n_estimators' :[1,5,10]
    }
    },
    'LR' :{
        'model' : LogisticRegression(solver='liblinear',multi_class='auto'),
        'pram' :{
            'C': [1,5,10]
        }
    },
   
    
}
```

```python
from sklearn.model_selection import GridSearchCV
score =[]

for model_name, model_pram in model_prams.items():
    clf = GridSearchCV(model_pram['model'], model_pram['pram'], cv = 5)
    clf.fit(X_train, y_train_for_Y_cor)
    score.append({
        'model': model_name,
        'best_score': clf.best_score_,
        'best_params': clf.best_params_
    })
```

```python
df_hypertunnig = pd.DataFrame(score, columns = ['model','best_score','best_params'])
```
